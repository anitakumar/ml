```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df = pd.read_excel('Social_Network_Ads.xlsx')
```

```python
df['Purchased'].value_counts()
```

```python
gender =pd.get_dummies(df['Gender'],drop_first=True)
df=pd.concat([df,gender],axis=1)
df.drop('Gender',axis=1,inplace=True)
```

```python
X=df.drop('Purchased',axis=1)
y=df['Purchased']
#
```

```python
# Import the StandardScaler class from the sklearn.preprocessing module.
from sklearn.preprocessing import StandardScaler

# Instantiate the StandardScaler object with the correct spelling.
scaler = StandardScaler()

# Fit and transform the data using the scaler object.
X = scaler.fit_transform(X)
```

```python
scaled_features=scaler.transform(X)
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features)
df_feat
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier # Import KNeighborsClassifier from sklearn.neighbors
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
                     metric_params=None, n_jobs=None, n_neighbors=5, p=2,
                     weights='uniform')
```

```python
pred=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(y_test,pred)
```
