```python
# prompt: write a function to generate =unique item i.e key form the dataset df

def generate_unique_items():
  unique_items_list = []
  for user_data in dataset.values():
    for item in user_data.keys():
      if item not in unique_items_list:
        unique_items_list.append(item)
  return unique_items_list


unique_items = generate_unique_items()
print(unique_items)
```

```python
def item_similarity(item1, item2):
  both_rated = {}
  for user in dataset.keys():
    if item1 in dataset[user] and item2 in dataset[user]:
      both_rated[user] = [dataset[user][item1], dataset[user][item2]]
  number_of_ratings = len(both_rated)
  if number_of_ratings == 0:
    return 0
  item1_ratings = [[dataset[user][item1] for user in both_rated]]
  item2_ratings = [[dataset[user][item2] for user in both_rated]]
  similarity = cosine_similarity(item1_ratings, item2_ratings)
  return similarity[0][0]
```

```python
print(item_similarity('Special Ops', 'Panchayat'))
```

```python
def most_similar_items(target_item):
  un_list =generate_unique_items() # Call the function generate_unique_items to get the list of unique items
  scores=[(item_similarity(target_item,other_item), target_item+"--> " +other_item) for other_item in un_list if other_item !=target_item]
  scores.sort(reverse=True)
  return scores
```

```python
print(most_similar_items('Special Ops'))
```
