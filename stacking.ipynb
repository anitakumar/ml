```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from vecstack import stacking
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
link='https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
names=['Class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
df=pd.read_csv(link,names=names)
df.head()
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#import XGBCLASSIFIER
from xgboost import XGBClassifier
#import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models=[GradientBoostingClassifier(),KNeighborsClassifier(),XGBClassifier(), RandomForestClassifier()]
```

```python
#import stacking
from sklearn.preprocessing import LabelEncoder
!pip install vecstack
from vecstack import stacking
from sklearn.metrics import accuracy_score # import the missing module
import warnings
le=LabelEncoder()
y_train=le.fit_transform(y_train)
S_train, S_test = stacking(models,X_train,y_train,X_test,regression=False,mode='oof_pred',needs_proba=False,save_dir=None,metric=accuracy_score,stratified=True,n_folds=4,shuffle=True,random_state=0,verbose=2)
warnings.filterwarnings('default',category=FutureWarning)
```
